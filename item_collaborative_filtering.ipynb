```python
import pandas as pd 
import numpy as np 
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
print(movies.shape)
print(ratings.shape)
```

```python
movies.head(3)
```

```python
ratings.head(3)
```

```python
df = pd.merge(ratings , movies , on = 'movieId')
df = df[['userId' , 'movieId' , 'rating' , 'title']]
print(df.head(3))
df_matrix = df.pivot_table('rating' , index = 'userId' , columns = 'title')
df_matrix = df_matrix.fillna(0)
df_matrix.head(3)
```

**item item collaborative filtering**

```python
item_matrix = df_matrix.transpose()
item_matrix.head(3)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
item_cos = cosine_similarity(item_matrix , item_matrix)
item_cos_df = pd.DataFrame(data = item_cos , index = df_matrix.columns , columns = df_matrix.columns)
item_cos_df.head(3)
```

```python
item_cos_df['Godfather, The (1972)'].sort_values(ascending=False)[:6]
```

```python
def predict_rating(ratings_matrix , item_cos_matrix):
    ratings_pred = ratings_matrix.dot(item_cos_matrix) / np.array([np.abs(item_cos_matrix).sum(axis = 1)])
    return ratings_pred 
 
ratings_pred = predict_rating(df_matrix , item_cos_df)
ratings_pred_df = pd.DataFrame(data = ratings_pred , index = df_matrix.index , columns = df_matrix.columns)
ratings_pred_df.head(3)
```

**Top-20 무비들로만 예측하기**

```python
def predict_ratings_top(ratings_arr, item_sim_arr , n = 20):
    pred = np.zeros(ratings_arr.shape)
    for col in range(ratings_arr.shape[1]):
        top_n_items = [np.argsort(item_sim_arr[: , col])[:-n-1:-1]]
        for row in range(ratings_arr.shape[0]):
            pred[row , col] = item_sim_arr[col , :][top_n_items].dot(ratings_arr[row, :][top_n_items].T)
            pred[row , col] /= np.sum(np.abs(item_sim_arr[col,:][top_n_items]))
    return pred

ratings_pred_df = predict_ratings_top(df_matrix.values , item_cos_df.values)
```

**사용자가 보지 않은 영화 추천하기(예 : 9 번 유저)**

```python
user_rating_id = df_matrix.loc[9 , :]
user_rating_id[user_rating_id > 0].sort_values(ascending = False)[:10]
```

```python
def get_unseen_movie(ratings_matrix , userId):
    user_rating = df_matrix.loc[userId , :]
    unseen_list = user_rating[user_rating == 0].index.tolist()
    return unseen_list

unseen_list = get_unseen_movie(df_matrix , 9)
unseen_list[:3]
```

```python
ratings_pred_df = pd.DataFrame(data = ratings_pred_df , index = df_matrix.index , columns = df_matrix.columns)
```

```python
def recomm_movie_by_userid(pred_df , userId , unseen_list , top_n = 10):
    recomm_movies = pred_df.loc[userId , unseen_list].sort_values(ascending = False)[:top_n]
    return recomm_movies

recomm_movie_by_userid(ratings_pred_df , 9 , unseen_list)
```
