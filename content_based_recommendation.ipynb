```python
import pandas as pd
import numpy as np
movies = pd.read_csv('tmdb_5000_movies.csv')
movies.head(5)
```

```python
movie_df = movies[['id' , 'title' , 'genres' , 'vote_average' , 'vote_count' , 'popularity' , 'keywords' , 'overview' ]]
movie_df.head(3)
```

```python
movie_df.info()
```

**literal_eval 을 사용해서 딕셔너리 형태로 바꾸기**

```python
from ast import literal_eval
movie_df['genres'] = movie_df['genres'].apply(literal_eval)
movie_df['keywords'] = movie_df['keywords'].apply(literal_eval)
movie_df['genres']
```

```python
movie_df['genres'] = movie_df['genres'].apply(lambda x : [y['name'] for y in x])
movie_df['keywords'] = movie_df['keywords'].apply(lambda x : [y['name'] for y in x])
movie_df[['genres' , 'keywords']]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
movie_df['genres_literal'] = movie_df['genres'].apply(lambda x : (' ').join(x))
movie_df['genres_literal']
```

```python
count_vect = CountVectorizer(min_df=0.0 , ngram_range = (1,2))
genre_mat = count_vect.fit_transform(movie_df['genres_literal'])
print(genre_mat.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
genre_cos = cosine_similarity(genre_mat , genre_mat)
```

```python
genre_cos_sort_index = genre_cos.argsort()[: , ::-1]
genre_cos_sort_index[1]
```

```python
def find_sim_movie(df , sort_index , title_name , top_n = 20):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    similar_indexs = sort_index[title_index , :top_n]
    similar_indexs = similar_indexs.reshape(-1)
    return df.iloc[similar_indexs]
    
```

```python
similar_df = find_sim_movie(movie_df , genre_cos_sort_index , 'The Godfather')
similar_df[['title' , 'vote_average']]
```

**하지만 장르 기반으로만 추천을 한다면 평점이 0.0과 같은 구린 영화를 추천하게 된다.**

```python
movie_df[['title' , 'vote_average' , 'vote_count']].sort_values('vote_average' , ascending = False)[:10]
```

**그래서 평점을 가중 해줘서 결과에 반영하게 만들 것임**

```python
C = movie_df['vote_average'].mean() #전체 영화의 평균 평점
m = movie_df['vote_count'].quantile(0.6) # 상위 60퍼센트 투표 횟수 평균
print("C :" , C)
print('m :' , m)
```

```python
def weighted_vote_average(record):
    V = record['vote_count']
    R = record['vote_average']
    return ((V / (V + m) * R) + ((m / (m + V)) * C))

movie_df['weighted_vote'] = movie_df.apply(weighted_vote_average , axis = 1)
```

```python
movie_df[['title' , 'vote_average' , 'vote_count' , 'weighted_vote']].sort_values('weighted_vote' , ascending = False)[:10]
```
